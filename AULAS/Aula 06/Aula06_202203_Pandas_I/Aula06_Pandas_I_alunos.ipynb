## O que é um DataFrame ?

O Pandas DataFrame é uma estrutura de dados tabular bidimensional de tamanho variável, podendo ou não ser heterogênea (vários tipos de dados), com eixos rotulados (linhas e colunas). O arranjo de um DataFrame permite a distribuição dos dados em linhas e colunas. Assim, podemos destacar três componentes presentes em um data frame: 
+ linhas
+ colunas
+ valores

## Criação de um Data Frame

Para inicializar um data fame utilizamos a função abaixo:

pandas.DataFrame(```data```, ```index```, ```columns```, ```dtype```)


* ```data```: numpy array, listas, series
* ```index```:  É opcional. Caso não seja informado, um objeto np.arange(m) será utilizado
* ```columns```: Também é opcional e é inicializado como np.arange(n) se não passado.
* ```dtype```: tipo de dado das colunas

## Transformando os dados

````python
df_pokemons[['Attack','Sp. Atk']].head()
````

````python
# Combinando os attacks e criando uma nova coluna
df_pokemons['Total_Attack'] = df_pokemons['Attack'] + df_pokemons['Sp. Atk']
df_pokemons[['Attack','Sp. Atk','Total_Attack']].head()
````

## Filtrando Dados

Podemos filtrar as linhas do data frame baseando-se em condições.

````python
#Uma condição
````

````python
#.loc
````

````python
# condições múltiplas ('&' equivale a 'and',  '|' equivale a 'or')
df_pokemons[(df_pokemons.HP>100) & (df_pokemons.Attack<80)].head()
````

````python
#.loc ('&' equivale a 'and',  '|' equivale a 'or')
df_pokemons.loc[(df_pokemons['HP']>100) | (df_pokemons['Attack']<80)].head()
````

## Mudança Condicional

Baseando-se nas condições que desejarmos, podemos substituir os valores de células.

Exemplo: Todo Type 1 de qualquer pokémon cujo HP seja maior que 60 será "LENDÁRIO"

````python
# Alterando um único valor
df_pokemons.loc[df_pokemons['HP'] > 60, 'Type 1'] = 'LENDARIO'

#df_pokemons.loc[df_pokemons['Type 1']=='LENDARIO'].head()
df_pokemons[df_pokemons.HP>60]['Type 1'].value_counts()
````

````python
# Alterando mais de 1 valor
df_pokemons.loc[df_pokemons['HP'] > 60, ['Type 1', 'Legendary']]= ['LENDARIO', True]
df_pokemons.loc[df_pokemons['HP'] > 60].head()
````
